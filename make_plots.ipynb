```python
performance = {30:[],60:[],90:[],120:[],210:[],365:[],730:[]}
with open('recurrent_dqn_performance.txt') as f:
    lines = f.readlines()
    for i in range(2,40,2):
        performance[30].append(lines[i].split()[1][:-1])
    for i in range(42,80,2):
        performance[60].append(lines[i].split()[1][:-1])
    for i in range(82,120,2):
        performance[90].append(lines[i].split()[1][:-1])
    for i in range(122,160,2):
        performance[120].append(lines[i].split()[1][:-1])
    for i in range(162,200,2):
        performance[210].append(lines[i].split()[1][:-1])
    for i in range(202,240,2):
        performance[365].append(lines[i].split()[1][:-1])
    for i in range(242,280,2):
        performance[730].append(lines[i].split()[1][:-1])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

performance = {k:np.array(v,dtype=np.float64) for k,v in performance.items()}
performance
```

```python
for k,v in performance.items():
    print(v.shape)
```

```python
df_names = ['tesla','eth','btc','google','amazon','apple','microsoft','netflix','spotify','paypal','nike','disney','amd','intel','nvidia','ibm','alibaba','uber','airbnb']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


for days, profits in performance.items():
    plt.figure(figsize=(18, 6))
    plt.rcParams['axes.facecolor'] = 'white'
    colors = []
    for profit in profits:
        if profit < 0.95:
            colors.append('red')
        elif profit >= 0.95 and profit < 1.0:
            colors.append('orange')
        elif profit >= 1.0:
            colors.append(sns.color_palette()[0])
        
    sns.barplot(y=list(profits), x=list(df_names), palette=colors)
    
    plt.axhline(y=1, color='green', linestyle='--')
    plt.axhline(y=0.95, color='yellow', linestyle='--')
    
    plt.title(f"Profit for {days} days interval")
    plt.xlabel("Company")
    plt.ylabel("Profit")
    plt.show()
```
